```python
import numpy as np
import pylab as pl
```

```python
# compute r_eff numerically
def reff_num(counts, values):
    # normalize
    counts = counts / counts.sum()
    # assume values is uniformly sample, 
    # otherwise we would have to include the d(values) in the sum
    # instead of canceling the d(values) scalar in the mom6/mom2 step
    mom6 = (counts*values**6).sum()
    mom2 = (counts*values**2).sum()
    return (mom6/mom2)**0.25
```

```python
# Fig 8 in Hong-Hsi Lee et al.
# Electron microscopy 3-dimensional segmentation and quantification of axonal dispersion and diameter distribution in mouse brain corpus callosum
# data recovered from plots
data_long = np.genfromtxt('/home/raid2/paquette/Downloads/longaxis.csv')
data_short = np.genfromtxt('/home/raid2/paquette/Downloads/shortaxis.csv')
```

```python
reff_short = reff_num(data_short[:,1]/data_short[:,1].sum(), data_short[:,0])
reff_long  = reff_num(data_long[:,1] /data_long[:,1].sum(),   data_long[:,0])
pl.figure()
c_long = 'blue'
c_short = 'red'
pl.plot(data_long[:,0],  data_long[:,1] /data_long[:,1].sum(),  color=c_long,  label='long  Axis, deff = {:.2f} um'.format(reff_long))
pl.plot(data_short[:,0], data_short[:,1]/data_short[:,1].sum(), color=c_short, label='short Axis, deff = {:.2f} um'.format(reff_short))
pl.axvline(reff_long, color=c_long)
pl.axvline(reff_short, color=c_short)
pl.legend()
pl.show()
```
